```python
def extract_match_details(index, url):
    # Fetch the HTML content
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Example: Extract match ID from a specific element (adjust as per your HTML structure)
    match_id = main_table.loc[index, 'Scorecard']

    # Extract all headings
    headings = soup.find_all('div', class_='ds-bg-ui-fill-translucent-hover')

    # Get the text of all headings
    headings_text = [heading.get_text(strip=True) for heading in headings]

    # Parse team names from headings
    team1 = headings_text[0].split('(')[0].strip() if len(headings_text) > 0 else None
    team2 = headings_text[1].split('(')[0].strip() if len(headings_text) > 1 else None

    # Extract the tables
    tables = soup.find_all('table')

    # Read the first and third tables into DataFrames
    df1 = pd.read_html(str(tables[0]))[0]
    df2 = pd.read_html(str(tables[2]))[0]

    # Remove the last four rows from each DataFrame
    df1 = df1.iloc[:-4]
    df2 = df2.iloc[:-4]

    # Remove the last two columns from each DataFrame
    df1 = df1.iloc[:, :-2]
    df2 = df2.iloc[:, :-2]

    # Remove rows where 'BATTING' column is NaN
    if 'BATTING' in df1.columns:
        df1 = df1.dropna(subset=['BATTING'])
    if 'BATTING' in df2.columns:
        df2 = df2.dropna(subset=['BATTING'])

    # Add team names to each DataFrame
    df1['Team'] = team1
    df2['Team'] = team2

    # Add match_id column
    df1['match_id'] = match_id
    df2['match_id'] = match_id

    # Add out/not_out column based on 'Unnamed: 1'
    df1['out/not_out'] = df1['Unnamed: 1'].apply(lambda x: 'not_out' if 'not out' in str(x).lower() else 'out')
    df2['out/not_out'] = df2['Unnamed: 1'].apply(lambda x: 'not_out' if 'not out' in str(x).lower() else 'out')

    # Add match column
    df1['match'] = team1 + ' vs ' + team2
    df2['match'] = team1 + ' vs ' + team2

    # Drop the 'Unnamed: 1' column
    df1.drop(columns=['Unnamed: 1'], inplace=True)
    df2.drop(columns=['Unnamed: 1'], inplace=True)

    # Drop the 'M' column if it exists
    if 'M' in df1.columns:
        df1.drop(columns=['M'], inplace=True)
    if 'M' in df2.columns:
        df2.drop(columns=['M'], inplace=True)

    # Reset index and add batting position starting from 1
    df1.reset_index(drop=True, inplace=True)
    df2.reset_index(drop=True, inplace=True)

    df1['battingPosition'] = df1.index + 1
    df2['battingPosition'] = df2.index + 1

    df1['BATTING'] = df1['BATTING'].str.replace(r'(c)', '', regex=False).str.replace('†', '', regex=False).str.strip()
    df2['BATTING'] = df2['BATTING'].str.replace(r'(c)', '', regex=False).str.replace('†', '', regex=False).str.strip()

    # Concatenate both DataFrames
    final_df = pd.concat([df1, df2], ignore_index=True)

    return final_df
```

```python
# Iterate over each link in main_table['Link']
dfs = []

for index, link in enumerate(main_table['Link']):
    try:
        # Apply extract_match_details function to each link
        df = extract_match_details(index, link)
        # Append the resulting DataFrame to the list
        dfs.append(df)
    except KeyError as e:
        print(f"KeyError occurred for link: {link}")
        raise e

# Concatenate all DataFrames in the list into a final DataFrame
final_df = pd.concat(dfs, ignore_index=True)
```

```python
def extract_bowling_details(index, url):
    # Fetch the HTML content
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Example: Extract match ID from a specific element (adjust as per your HTML structure)
    match_id = main_table.loc[index, 'Scorecard']

    # Extract all headings
    headings = soup.find_all('div', class_='ds-bg-ui-fill-translucent-hover')

    # Get the text of all headings
    headings_text = [heading.get_text(strip=True) for heading in headings]

    # Parse team names from headings
    team1 = headings_text[0].split('(')[0].strip() if len(headings_text) > 0 else None
    team2 = headings_text[1].split('(')[0].strip() if len(headings_text) > 1 else None

    # Extract the tables
    tables = soup.find_all('table')

    # Read the second and fourth tables into DataFrames
    df1 = pd.read_html(str(tables[1]))[0]
    df2 = pd.read_html(str(tables[3]))[0]


    # Remove rows where 'BOWLING' column starts with a digit
    df1 = df1[~df1['BOWLING'].str.match(r'^\d')]
    df2 = df2[~df2['BOWLING'].str.match(r'^\d')]

    # Add team names to each DataFrame
    df1['Team'] = team2
    df2['Team'] = team1

    # Add match_id column
    df1['match_id'] = match_id
    df2['match_id'] = match_id

    # Add match column
    df1['match'] = team1 + ' vs ' + team2
    df2['match'] = team1 + ' vs ' + team2

    # Reset index and add bowling position starting from 1
    df1.reset_index(drop=True, inplace=True)
    df2.reset_index(drop=True, inplace=True)


    # Concatenate both DataFrames
    final_df = pd.concat([df1, df2], ignore_index=True)

    return final_df
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.espncricinfo.com/series/icc-men-s-t20-world-cup-2024-1411166/afghanistan-squad-1431702/series-squads"

# Fetch the HTML content from the URL
response = requests.get(url)
html_content = response.content

# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Search for elements with the specific classes
player_elements = soup.find_all(class_="ds-border-line odd:ds-border-r ds-border-b")

# Extract information from each player element
for player_elem in player_elements:
    # Extracting player name
    player_name_elem = player_elem.find('a', class_='ds-inline-flex ds-items-start ds-leading-none')
    player_name = player_name_elem.text.strip() if player_name_elem else ""

    # Extracting player role
    player_role_elem = player_elem.find('p', class_='ds-text-tight-s ds-font-regular ds-mb-2 ds-mt-1')
    player_role = player_role_elem.text.strip() if player_role_elem else ""

    # Extracting age, batting style, and bowling style
    details_div = player_elem.find('div', class_='ds-justify-between ds-text-typo-mid3')
    details = details_div.find_all('div', class_='ds-flex ds-items-start ds-space-x-1')

    # Extract age, batting style, and bowling style
    age = details_div.find('div', class_='ds-flex ds-items-center ds-space-x-1').find('span', class_='ds-text-compact-xxs ds-font-bold').text.strip()
    batting_style = ""
    bowling_style = "None"

    # Loop through the details to extract batting and bowling styles
    for detail in details:
        if 'Batting:' in detail.text:
            batting_style = detail.find('span', class_='ds-text-compact-xxs ds-font-bold').text.strip()
        elif 'Bowling:' in detail.text:
            bowling_style = detail.find('span', class_='ds-text-compact-xxs ds-font-bold').text.strip()

    # Print the extracted information
    print(f"Name: {player_name}")
    print(f"Age: {age}")
    print(f"Batting Style: {batting_style}")
    print(f"Player Role: {player_role}")
    print(f"Bowling Style: {bowling_style}")
    print("-" * 30)
```
