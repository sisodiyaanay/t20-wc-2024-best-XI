```python
import requests
from bs4 import BeautifulSoup

url = "https://www.espncricinfo.com/series/icc-men-s-t20-world-cup-2024-1411166/afghanistan-squad-1431702/series-squads"

response = requests.get(url)
html_content = response.content

soup = BeautifulSoup(html_content, 'html.parser')

player_elements = soup.find_all(class_="ds-border-line odd:ds-border-r ds-border-b")

for player_elem in player_elements:
    player_name_elem = player_elem.find('a', class_='ds-inline-flex ds-items-start ds-leading-none')
    player_name = player_name_elem.text.strip() if player_name_elem else ""

    player_role_elem = player_elem.find('p', class_='ds-text-tight-s ds-font-regular ds-mb-2 ds-mt-1')
    player_role = player_role_elem.text.strip() if player_role_elem else ""

    batting_style_elem = player_elem.find('div', class_='ds-flex ds-items-start ds-space-x-1').find('span', text='Batting:')
    batting_style = batting_style_elem.find_next_sibling('span').text.strip() if batting_style_elem else ""

    bowling_style_elem = player_elem.find('div', class_='ds-flex ds-items-start ds-space-x-1').find('span', text='Bowling:')
    bowling_style = bowling_style_elem.find_next_sibling('span').text.strip() if bowling_style_elem else "None"

    print(f"Name: {player_name}")
    print(f"Batting Style: {batting_style}")
    print(f"Player Role: {player_role}")
    print(f"Bowling Style: {bowling_style}")
    print("-" * 30)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.espncricinfo.com/series/icc-men-s-t20-world-cup-2024-1411166/afghanistan-squad-1431702/series-squads"

response = requests.get(url)
html_content = response.content

soup = BeautifulSoup(html_content, 'html.parser')

first_ds_p_0 = soup.find(class_="ds-p-0")

base_url = "https://www.espncricinfo.com"
teams = []
links = []
if first_ds_p_0:
    a_tags = first_ds_p_0.find_all('a')
    for a_tag in a_tags:
        if 'href' in a_tag.attrs:
            full_url = base_url + a_tag['href']
            team_name = a_tag.text.strip()
            links.append(full_url)
            teams.append(team_name)

df = pd.DataFrame({
    'Team Name': teams,
    'Link': links
})

df
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def extract_player_info(url, team_name):
    response = requests.get(url)
    html_content = response.content

    soup = BeautifulSoup(html_content, 'html.parser')

    player_elements = soup.find_all(class_="ds-border-line odd:ds-border-r ds-border-b")

    player_data = []
    for player_elem in player_elements:
        player_name_elem = player_elem.find('a', class_='ds-inline-flex ds-items-start ds-leading-none')
        player_name = player_name_elem.text.strip() if player_name_elem else ""

        player_role_elem = player_elem.find('p', class_='ds-text-tight-s ds-font-regular ds-mb-2 ds-mt-1')
        player_role = player_role_elem.text.strip() if player_role_elem else ""

        details_div = player_elem.find('div', class_='ds-justify-between ds-text-typo-mid3')
        details = details_div.find_all('div', class_='ds-flex ds-items-start ds-space-x-1')

        age = details_div.find('div', class_='ds-flex ds-items-center ds-space-x-1').find('span', class_='ds-text-compact-xxs ds-font-bold').text.strip()
        batting_style = ""
        bowling_style = "None"

        for detail in details:
            if 'Batting:' in detail.text:
                batting_style = detail.find('span', class_='ds-text-compact-xxs ds-font-bold').text.strip()
            elif 'Bowling:' in detail.text:
                bowling_style = detail.find('span', class_='ds-text-compact-xxs ds-font-bold').text.strip()

        player_data.append({
            "Name": player_name,
            "Team": team_name,
            "BattingStyle": batting_style,
            "BowlingStyle": bowling_style,
            "PlayingRole": player_role
        })

    return player_data

data = df
df_new = pd.DataFrame(data)

all_players_data = []

for index, row in df_new.iterrows():
    url = row['Link']
    team_name = row['Team Name']

    players_data = extract_player_info(url, team_name)

    all_players_data.extend(players_data)

players_df = pd.DataFrame(all_players_data)

players_df
```

```python
players_df.to_csv('players_info.csv', index=False)
```
